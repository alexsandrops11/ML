```python
# Bibliotecas
import pandas as pd
import os
import numpy as np
from PIL import Image, ImageFilter
import glob
import pickle
import csv
from keras.models import load_model
```

```python
doen = ['akiec', 'bcc', 'mel', 'nv', 'bkl', 'df', 'vasc']
path = 'D:/cancer de pele/data_teste/teste.csv'
dados = pd.read_csv(path)
dados.head(5)
```

```python
i = 15
j = 10
path1 = 'D:/cancer de pele/'
ModelCancer = load_model(path1 + 'modelo/' + str(i) + '/ModelCancer-' + str(j) + '.h5')
```

```python
os.chdir(path1 + 'data_teste/')
arq = glob.glob('*.pckl')
target = []
cont = 0



for j in arq:
    x = [0]*10
    imgF = np.zeros((1,250,250,3))
    
    x[0] = j[:-4]
    
    x[8] = doen[dados.loc[dados['image_id'] == j[:-5]].values.tolist()[0][1:].index(1)]
    
    
    name = j[:-5] + '.pckl'
    f = open(name, 'rb')
    xt = pickle.load(f)
    
    imgF[0,:,:,:] = xt
    
    predi = ModelCancer.predict(imgF[:1,:,:,:])[0]
    x[9] = doen[np.argmax(predi)]
    
    for i in range(7):
        x[i+1] = predi[i]
    
    target.append(x)
    
```
