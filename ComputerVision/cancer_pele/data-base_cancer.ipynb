```python
doen = []
doen.append(np.count_nonzero(dados['dx'] == 'akiec'))
doen.append(np.count_nonzero(dados['dx'] == 'bcc'))
doen.append(np.count_nonzero(dados['dx'] == 'mel'))
doen.append(np.count_nonzero(dados['dx'] == 'nv'))
doen.append(np.count_nonzero(dados['dx'] == 'bkl'))
doen.append(np.count_nonzero(dados['dx'] == 'df'))
doen.append(np.count_nonzero(dados['dx'] == 'vasc'))
print(doen)
print(sum(doen))
```
