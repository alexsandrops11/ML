```python
# Bibliotecas
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageFilter
import glob
from numpy import random as rd
import pickle
import csv
from keras.models import load_model
```

```python
path0="E:/cancer de pele/data_valida/"
os.chdir(path0)
files=glob.glob("*.pckl")
Path     =  'E:/cancer de pele/validação/0-validacao.pckl'
```

```python
dados = pd.read_csv(path0 + 'validacao.csv')

dados.head(10)
```

# CRIAR O PACK DE VALIDATION

```python
size = len(files)
X = np.zeros((size,250,250,3))
Y = [0]*size

i = 0
for a in files:
    #ABRIR IMAGENS 'A' PICKLE
    f = open(a, 'rb')
    X[i] = pickle.load(f)
    

    #ACHAR O RESPECTIVO LABEL DA IMAGEM 'A' ACHADA NO DATASET TARGET CRIADO ANTERIOREMENTE
    x = dados.loc[dados['image_id'] == a[:-5]]
    lol = x.values.tolist()
    Y[i] = lol[0][1:]
    i+=1
    if i%100 == 0:
        print(i)

## Save images and labels in pckl format
f = open(Path, 'wb')
pickle.dump([X,Y], f)
f.close()
print('TERMINOU')
```

# RODAR A PARTIR DAQUI PARA VALIDAÇÃO

```python
f = open(Path, 'rb')
[Xt,Yt] = pickle.load(f)
Yt = np.array(Yt)
```

```python
for i in range(24,25):
    result = []
    for j in range(11):
        #load model
        model_path = 'E:/cancer de pele/modelo/' + str(i) + '/ModelCancer-' + str(j) + '.h5'
        ModelCancer = load_model(model_path)

        ## Model validation
        preds = ModelCancer.evaluate(Xt, Yt)
        print ("Loss = " + str(round(preds[0],3)))
        print ("Test Accuracy = " + str(round(preds[1],3)))

        #salvar resultados
        result.append([ i , j ,preds[0], preds[1]])
        dats = pd.DataFrame(result, columns= ["id","model","loss", "accuracy"])
        dats.to_csv('e:/cancer de pele/modelo/' + str(i) + '/RESULT_VALIDACAO.csv')
    print('ESTA NO TREINO :' + str(i))
```
