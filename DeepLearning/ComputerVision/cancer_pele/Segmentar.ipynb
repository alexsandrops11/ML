SEGMENTAR O BANCO DE DADOS EM AMOSTRAS RANDOMICAS DE TAMANHO 'SIZE'

```python
pasta = 27
size = 809
times = len(files)//size

    
dados = pd.read_csv(path0 + 'treino_target.csv')
dados.drop('Unnamed: 0', axis = 1, inplace = True)
dados.head(10)

for run in range(times):

    X = np.zeros((size,250,250,3))
    Y = [0]*size

    #INDICES DE FILES RANDOMICOS
    imagens = rd.choice(range(len(files)), size, replace = False) # comando de vetor aleatório, size é o número de imagens

    #ORGANIZAR DE FORMA DECRESCENTE PARA NO .POP RETIRAR PRIMEIRO OS MAIS ALTOS  NÃO ALTERANDO O INDICES DOS PROXIMOS
    imagens = sorted(imagens, key=int,   reverse=True)
    i = 0
    for a in imagens:
        #ABRIR IMAGENS 'A' PICKLE
        f = open(files[a], 'rb')
        X[i] = pickle.load(f)

        #ACHAR O RESPECTIVO LABEL DA IMAGEM 'A' ACHADA NO DATASET TARGET CRIADO ANTERIOREMENTE
        x = dados.loc[dados['image_id'] == files[a][:-5]]
        lol = x.values.tolist()
        Y[i] = lol[0][1:]
        i+=1
        if i%100 == 0:
            print(i)

        #DELETAR IMAGEM 'A' DO FILES PARA NÃO REPETIR A SELEÇÃO DE IMAGENS
        files.pop(a)

    ## Save images and labels in pckl format
    Path     =  'E:/cancer de pele/TRAIN/'+ str(pasta) + '/' + str(run) + '-Data.pckl'
    f = open(Path, 'wb')
    pickle.dump([X,Y], f)
    f.close()
    print('Está na vez: ' + str(run+1) + ' de ' + str(times))
```
